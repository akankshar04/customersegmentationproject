```python
age_group1 = data.where('Customer Age', are.between(20, 30)).group('Product Category')
age_group2 = data.where('Customer Age', are.between(30, 40)).group('Product Category')
age_group3 = data.where('Customer Age', are.between(40, 50)).group('Product Category')
age_group1.barh('Product Category')
age_group2.barh('Product Category')
age_group3.barh('Product Category')
```

```python
age_group4 = data.where('Customer Age', are.between(50, 60)).group('Product Category')
age_group5 = data.where('Customer Age', are.between(60, 70)).group('Product Category')
age_group6 = data.where('Customer Age', are.between(70, 80)).group('Product Category')
age_group4.barh('Product Category')
age_group5.barh('Product Category')
age_group6.barh('Product Category')
```

```python
#lets look at the accesories across the ages in terms of percenatages
```

```python
accessories_1 = age_group1.with_columns('percentage', age_group1.column('count')/sum(age_group1.column('count'))).take(0).select('count', 'percentage')
accessories_2 = age_group2.with_columns('percentage', age_group2.column('count')/sum(age_group2.column('count'))).take(0).select('count', 'percentage')    
accessories_3 = age_group3.with_columns('percentage', age_group3.column('count')/sum(age_group3.column('count'))).take(0).select('count', 'percentage')
accessories_4 = age_group4.with_columns('percentage', age_group4.column('count')/sum(age_group4.column('count'))).take(0).select('count', 'percentage') 
accessories_5 = age_group5.with_columns('percentage', age_group5.column('count')/sum(age_group5.column('count'))).take(0).select('count', 'percentage')
accessories_6 = age_group6.with_columns('percentage', age_group6.column('count')/sum(age_group6.column('count'))).take(0).select('count', 'percentage') 


accessories_1.show()
accessories_2.show()
accessories_3.show()
accessories_4.show()
accessories_5.show()
accessories_6.show()
```

```python
bikes_1 = age_group1.with_columns('percentage', age_group1.column('count')/sum(age_group1.column('count'))).take(1).select('count', 'percentage')
bikes_2 = age_group2.with_columns('percentage', age_group2.column('count')/sum(age_group2.column('count'))).take(1).select('count', 'percentage')
bikes_3 = age_group3.with_columns('percentage', age_group3.column('count')/sum(age_group3.column('count'))).take(1).select('count', 'percentage')
bikes_4 = age_group4.with_columns('percentage', age_group4.column('count')/sum(age_group4.column('count'))).take(1).select('count', 'percentage')
bikes_5 = age_group5.with_columns('percentage', age_group5.column('count')/sum(age_group5.column('count'))).take(1).select('count', 'percentage')
bikes_6 = age_group6.with_columns('percentage', age_group6.column('count')/sum(age_group6.column('count'))).take(1).select('count', 'percentage')


bikes_1.show()
bikes_2.show()
bikes_3.show()
bikes_4.show()
bikes_5.show()
bikes_6.show()
```

```python
#what products are most popular?
#lets add a profitability column 
```

```python
data_w_profitability= data.with_columns('profit_percentage', (data.column('Unit Price')-data.column('Unit Cost'))
                                        /data.column('Unit Price'))
```

```python
data_w_profitability
```

```python
data_w_profitability.group('Product Category', np.average).sort('profit_percentage average', descending=True)
```

```python
#The accesories category has the highest average profitability
```

```python
data_w_profitability.group('Sub Category', np.average).sort('profit_percentage average', descending=True)
```

```python
#here we can see that the average profit percentage is highest in the Bike racks, Fenders, gloves sub categories
```
